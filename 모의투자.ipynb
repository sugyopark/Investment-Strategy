# 간단한 모의투자 전략

```python
import importlib
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import yfinance as yf
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 데이터를 불러옵니다.
price = pd.read_csv("수정주가.csv", index_col = 0)
volume = pd.read_csv("시가총액.csv", index_col=0)
institution = pd.read_csv("기관계.csv", index_col=0, encoding = 'cp949')
per = pd.read_csv("per.csv", index_col=0)
ticker = pd.read_csv("KOSPI_ticker.csv", index_col=0)
```

```python
def convert_to_numeric(df):
    for column in df.columns:
        # 데이터가 문자열 형식인지 확인
        if df[column].dtype == 'object':
            # 쉼표를 제거하고 숫자형으로 변환
            df[column] = pd.to_numeric(df[column].str.replace(',', ''), errors='coerce')
    return df
```

```python
def filter_trade(kospi_trade, kospi_institution):
    # 날짜 열을 datetime 객체로 변환
    kospi_trade.index = pd.to_datetime(kospi_trade.index)
    kospi_institution.index = pd.to_datetime(kospi_institution.index)

    # 두 번째 데이터 프레임의 날짜 범위 확인
    start_date = kospi_institution.index.min()
    end_date = kospi_institution.index.max()

    # 첫 번째 데이터 프레임에서 해당 범위의 데이터 선택
    kospi_trade_filtered = kospi_trade[(kospi_trade.index >= start_date) & (kospi_trade.index <= end_date)]

    kospi_trade_filtered_convert = convert_to_numeric(kospi_trade_filtered)
    kospi_institution_convert = convert_to_numeric(kospi_institution)
    result_df = kospi_institution_convert.div(kospi_trade_filtered_convert)

    return result_df
```

## 1. 데이터 전처리

```python
# csv file이므로 전처리를 진행해줍니다.
kospi_price = convert_to_numeric(price)
per = convert_to_numeric(per)
```

```python
trade = filter_trade(volume, institution)
#trade = trade.fillna(0)
trade
```

```python
# 각 회사별로 3개월 간 주가 데이터를 계산
price_60_days_ago = kospi_price.shift(60)
price_1_days_ago = kospi_price.shift(1)

# 변화율 계산
change_rate = ((price_1_days_ago - price_60_days_ago) / price_1_days_ago) * 100

# 결과 
display(change_rate)
```

## 2. 스크리닝 (투자할 기업 선정)

```python
# 지난 20일 동안의 누적 순매수 계산
cumulative_trade = trade.rolling(window=60).sum()

# 결과 출력
trade_list = cumulative_trade.iloc[-1, :].sort_values()
trade_list = trade_list[:len(trade_list)//10].index # 순매도 상위 10%
trade_list
```

```python
# 주가 수익률 상위 10% 기업 screening
change_list = change_rate.iloc[-1, :].sort_values(ascending = False)
change_list = change_list[:len(change_list)//10].index
change_list
```

```python
# 두 개의 리스트의 교집합을 구합니다. (기관 순매도 상위 10% & 주가 수익률 상위 10%)
filter = trade_list.intersection(change_list)
filter
```

```python
# 현재 해당하는 기업을 뽑습니다.
trade_present = per.loc[: , filter].iloc[-1, :].sort_values()[:10]
ticker.loc[trade_present.index, :]
```

```python
# 리밸런싱 조건 설정해주기
# 리밸런싱 주기
rebalancing = 60 #일 단위

# 거래비용
tax = 0.0002 # 현재 거래비용은 0.02%입니다.

# 투자원금
cash = 100000000 # 모의투자 시작금액인 1억으로 시작하겠습니다.
money = cash

risk_free_rate = 0.0359 # 23년 4월 1일 기준 CD91일물 금리로 하겠습니다.

# 손절기준
loss_cut = 0.1 #매수 이후 10% 이상 가격이 떨어지면 매도하겠습니다.

day = 0
quarter = 0
df_list = '현금 주식총액 포트폴리오가치 일일수익률 총수익률'.split() # 백테스트 데이터프레임에 보여줄 항목들입니다. 이후에 더 추가할거에요
backtest = pd.DataFrame(columns = df_list)
```

## 3. 백테스트 (전략 검증)

```python
print('전체 리밸런싱 횟수는 {}'.format(int(len(kospi_price)/rebalancing)))
for reb in range(int(len(kospi_price)/rebalancing)): #리밸런싱 횟수를 계산해줍니다. 전체 투자기간/리밸런싱 주기
    print( reb+1, '회 투자')
    print('현금 : ', money)
    inv = kospi_price.iloc[day:day+rebalancing,:] #전체 주가 데이터에서 투자기간동안의 데이터를 뽑아냅니다.
    inv = inv.dropna(axis=1) # 투자 시작 시점에 데이터가 Nan인 친구들을 날려줍니다.

    # change_rate로 20일 주가수익률 기준으로 필터
    change_inv = change_rate.iloc[day:day+rebalancing, :]
    change_list = change_inv.iloc[0, :].sort_values(ascending = False)
    change_list = change_list[:len(change_inv)//10].index # 주가 수익률 상위 10% 기업
    
    # 20일 누적순매수 계산(기관계)
    trade_inv = cumulative_trade.iloc[day:day+rebalancing, :]
    trade_list = trade_inv.iloc[0, :].sort_values()
    trade_list = trade_list[:len(trade_list)//10].index # 기관 순매수 하위 10% 기업

    filter_list = trade_list.intersection(change_list)
    print(filter_list)
    # per 랭킹해서 10개 필터
    per_inv = per.iloc[day:day+rebalancing, :]
    per_list = per_inv.loc[: , filter_list].iloc[0, :]
    per_list = per_list.sort_values() # 오름차순으로 정렬
    per_list = per_list[:10]
    inv_list = per_list.index # inv_list에 per_list의 index(filter + per 기준 value 높은 10개 기업 선정)
    #inv_list = filter_list
    print('투자 후보 갯수는 : ', len(inv_list))

    #여기까지 스크리닝 과정입니다.

    final_inv_list = []
    for i in range(len(inv_list)):
        if inv_list[i] in inv.columns:
            final_inv_list.append(inv_list[i])
        else:
            print(inv_list[i],' 종목이 없습니다')
    print('투자하는 종목의 수는 : ', len(final_inv_list))
    print('투자종목 : ',final_inv_list)
    # 스크리닝 된 종목 들 중 가격 데이터에 있는 종목들만 투자 리스트에 추가해줍니다.


    # 매수 기준 : 동일 비중
    # 보유한 주식의 가치 평가하기

    if len(final_inv_list)==0:
        allocation=0
    else:
        allocation = money / 10 # 동일 비중으로 진행하면 10개 종목이 아닌 경우도 있으므로 10으로 나눠줍니다.
    print('동일 비중 투자 금액은 : ' , allocation) # 저는 제 코드를 믿지 못하기 때문에 이렇게 중간중간 어디서 에러가 나는지 예상할 수 있게 출력 결과도 함께 프린트합니다.
    final_kospi_price = inv[final_inv_list].copy()



    vec = pd.DataFrame({'매수수량' : allocation // final_kospi_price.iloc[0,:]}) # 종목당 매수할 수를 데이터프레임(벡터) 로 만들어줍니다.
    vec = vec.replace(np.NaN, 0)

    # 매도(손절 기준) : 10% 이상 떨어질 경우 매도

    # 손절을 할 때 보유하고 있는 주식수를 0으로 만들거나
    # 손절 이후 포트폴리오의 가격을 0으로 만들거나의 방법을 선택할 수 있겠죠?
    # 저는 손절 한 순간부터 포트폴리오에서 주가를 계산하지 않아도 되는 방법이 더 편할 것 같아서 이렇게 선택했지만
    # 저와 다르게 계산하셔도 상관없습니다.
    loss_cut_money_list = []
    loss_cut_money = 0
    for days in final_kospi_price.index:
        for stocks in final_kospi_price.columns:
            if final_kospi_price.loc[days, stocks] < final_kospi_price[stocks][0] * ( 1 - loss_cut ):
                loss_cut_money = loss_cut_money + (final_kospi_price.loc[days, stocks] * float(vec.loc[stocks])) * (1 - tax)
                final_kospi_price.loc[days:, stocks] = 0

        loss_cut_money_list.append(loss_cut_money)


        # 날짜에 저장된 값을 리스트에 추가해줍니다. 이 반복문을 돌면서 포트폴리오 투자기간동안의 현금이 저장될 것입니다.



    product = np.dot(final_kospi_price, vec) #np.dot 통해 행렬과 벡터의 내적값을 구해줍니다.
    product = pd.DataFrame(product) #데이터프레임으로 만들어서 나중에 붙여주겠습니다.



    balance = pd.DataFrame(index = product.index)
    balance['현금'] = money - product.iloc[0,0] # 현금은 리밸런싱이 이루어지는 날의 현금에서 주식 매수대금을 뺀 값입니다.
    balance['현금'] += loss_cut_money_list # 손절한 금액은 날짜별로 더해줍시다


    _backtest = pd.DataFrame(columns = df_list) # 빈 데이터프레임을 만들어줍니다.
    _backtest['주식총액'] = product # 보유하고 있는 주식의 가치입니다.
    _backtest['현금'] = balance['현금'] # 아까 계산한 일자별 현금 데이터를 넣어줍니다.


    _backtest['포트폴리오가치'] = _backtest['현금'] + _backtest['주식총액'] # 포트폴리오 가치는 현금과 주식총액의 합입니다.
    _backtest['총수익률'] = (_backtest['포트폴리오가치']/cash) # 처음 들고 시작한 돈 대비 수익률
    _backtest.index = kospi_price.index[day : day + rebalancing]


    backtest = pd.concat([backtest, _backtest], axis = 0, ignore_index=False)
    # 데이터프레임들을 계속 붙여주면서 백테스트 데이터를 만들겠습니다.
    # 반복문을 돌면서 빈데이터프레임에 데이터를 채워가기 때문에 먼저 선언해준 백테스트 데이터프레임에 붙여줍니다.


    # 리밸런싱 이후 시작 금액은 주식 판매대금 - 거래비용 + 현금입니다.
    money = backtest.iloc[-1,0] + (backtest.iloc[-1,1] * ( 1 - tax))
    day = day + rebalancing
    quarter += rebalancing

backtest['일일수익률'] = backtest['포트폴리오가치'].pct_change()
```

```python
# Yahoo Finance library를 통해 코스피 주가 데이터를 받아옵니다.
BM=yf.download("^KS11",start="2014-01-01",end="2024-03-18")
BM
```

```python
# 유니버스가 코스피이므로, 코스피 지수를 벤치마크로 설정합니다.
BM['일일수익률'] = BM['Open'].pct_change()
BM['총수익률'] = BM['Open']/BM['Open'][0]
BM
```

```python
# CAGR : 연평균 성장률 ( Compound Annual Growth Rate)를 계산합니다.
num_of_year = float(len(backtest)/240)

CAGR = (backtest['포트폴리오가치'][-1]/cash)**(1/num_of_year) - 1
print('포트폴리오 CAGR은 {} %'.format(CAGR*100))

BENCHMARK_CAGR = ((BM['Open'][len(BM)-1] / BM['Open'][0])) ** (1 / num_of_year) - 1
print("벤치마크 CAGR은 {} %".format(BENCHMARK_CAGR*100))
```

```python
#MDD를 계산합니다.

max_list = [backtest.iloc[0,2]]
min_list = [backtest.iloc[0,2]]

for i in range(len(backtest)):
    if i==0:
        max_list.append(backtest.iloc[0,2])
        min_list.append(backtest.iloc[0,2])

    else:
        if backtest.iloc[i, 2] >= backtest.iloc[i-1,2]:
            max_list.append(backtest.iloc[i, 2])
            min_list.append(backtest.iloc[i, 2])
        else:
            if(max_list[-1]>backtest.iloc[:i,2].max()):
                max_list.append(max_list[-1])
            else:
                max_list.append(backtest.iloc[:i,2].max())
            min_list.append(backtest.iloc[i, 2])


max_list = max_list[1:]
min_list = min_list[1:]
backtest['max'] = max_list
backtest['min'] = min_list
backtest['mdd'] = -((backtest['max'] - backtest['min'])/backtest['max'])

print('BACKTEST MDD는 {}%'.format(backtest['mdd'].min()*100))

backtest['MDD'] = 0
for i in range(len(backtest)):
    if i != 0:
        backtest.iloc[i, -1] = backtest['mdd'][:i].min()
```

```python
#MDD를 계산합니다.

max_list = [BM.iloc[0,2]]
min_list = [BM.iloc[0,2]]

for i in range(len(BM)):
    if i==0:
        max_list.append(BM.iloc[0,2])
        min_list.append(BM.iloc[0,2])

    else:
        if BM.iloc[i, 2] >= BM.iloc[i-1,2]:
            max_list.append(BM.iloc[i, 2])
            min_list.append(BM.iloc[i, 2])
        else:
            if(max_list[-1]>BM.iloc[:i,2].max()):
                max_list.append(max_list[-1])
            else:
                max_list.append(BM.iloc[:i,2].max())
            min_list.append(BM.iloc[i, 2])


max_list = max_list[1:]
min_list = min_list[1:]
BM['max'] = max_list
BM['min'] = min_list
BM['mdd'] = -((BM['max'] - BM['min'])/BM['max'])

print('Benchmark MDD는 {}%'.format(BM['mdd'].min()*100))

BM['MDD'] = 0
for i in range(len(BM)):
    if i != 0:
        BM.iloc[i, -1] = BM['mdd'][:i].min()
```

```python
# Sharpe Ratio(사후적)= (자산 X의 수익률 – 무위험 자산 수익률) / 자산 X의 수익률의 표준편차
backtest_return = backtest['일일수익률'].mean()
backtest_std = backtest['일일수익률'].std()
BM_return = BM['일일수익률'].mean()
BM_std = BM['일일수익률'].std()

backtest_sharpe = (backtest_return - risk_free_rate/365) / backtest_std *np.sqrt(len(backtest))
BM_sharpe = (BM_return - risk_free_rate/365) / BM_std *np.sqrt(len(backtest))

print('BACKTEST 샤프 비율은 : ', backtest_sharpe)
print('BENCHMARK 샤프 비율은 : ', BM_sharpe)
```

```python
# 일일수익률의 평균과 표준편차 계산
backtest_return = backtest['일일수익률'].mean()
backtest_std = backtest['일일수익률'].std()
BM_return = BM['일일수익률'].mean()
BM_std = BM['일일수익률'].std()

# 연율화 계산
trading_days = 252  # 일반적으로 사용되는 연간 거래일 수
backtest_return_annualized = backtest_return * trading_days
BM_return_annualized = BM_return * trading_days

backtest_std_annualized = backtest_std * np.sqrt(trading_days)
BM_std_annualized = BM_std * np.sqrt(trading_days)

# 연율화된 샤프 비율 계산
backtest_sharpe_annualized = (backtest_return_annualized - risk_free_rate) / backtest_std_annualized
BM_sharpe_annualized = (BM_return_annualized - risk_free_rate) / BM_std_annualized

backtest_sharpe_annualized, BM_sharpe_annualized
```
